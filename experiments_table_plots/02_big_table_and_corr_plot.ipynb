```python
import pickle
import os

files = os.listdir("../results/base/")
base = [f for f in files if "base" in f]

result_dict = dict()

for file in base:
    with open("../results/base/"+file, "rb") as f:
        file_name = file.replace("base_","").replace(".pickle","")
        dataset, model = file_name.rsplit("_",1)

        print(dataset, model)
        if dataset not in result_dict.keys():
            result_dict[dataset] = dict()
        result_dict[dataset][model] = pickle.load(f)
```

```python
from collections import OrderedDict
import pandas as pd
import pickle
import numpy as np
import re
from datetime import datetime

print(pd.__version__)

def flatten_dict(nested_dict):
    res = OrderedDict()
    if isinstance(nested_dict, dict):
        for k in nested_dict:
            if k in ["best_valid_score", "valid_score_bigger", "best_valid_result"]:
                continue
            flattened_dict = flatten_dict(nested_dict[k])
            for key, val in flattened_dict.items():
                key = list(key)
                key.insert(-1, k)
                res[tuple(key)] = val
    else:
        res[()] = nested_dict
    return res

now = datetime.now()
time = str(now.strftime("%Y-%m-%d_%H%M%S"))

data = result_dict
```

```python
def highlight_max(x):
    return np.where(x == np.nanmax(x.to_numpy()), f"font-weight: bold;", None)

def highlight_min(x):
    return np.where(x == np.nanmin(x.to_numpy()), f"font-weight: bold;", None)

def add_cline(measure, latex_code, df):
    end_index = []
    for el in re.finditer(f"{measure}.*\\n",latex_code):
        end_index.append(el.end())
    j = df.shape[1] + 2
    for idx in reversed(end_index):
        latex_code= latex_code[:idx] + "\\cline{2-" + str(j) + "}\n" + latex_code[idx:]
    return latex_code

def nicetable(df, color=True):
    row_with_up = df.index.levels[1][df.index.levels[1].str.contains("uparrow")]
    row_with_down = df.index.levels[1][df.index.levels[1].str.contains("downarrow")]

    idx = pd.IndexSlice

    slice_max = idx[idx[:,row_with_up],:]
    slice_min = idx[idx[:,row_with_down],:]

    styler = df.style

    styler.apply(highlight_max, axis=1, subset=slice_max)\
        .apply(highlight_min, axis=1, subset=slice_min)
    
    if color:
        styler\
            .background_gradient(cmap="RdYlGn", axis=1, vmin=0, vmax=1, subset=slice_max)\
            .background_gradient(cmap="RdYlGn_r", axis=1, vmin=0, vmax=1, subset=slice_min)\

    latex_code = styler.to_latex(
        hrules=True, 
        clines="skip-last;data",
        convert_css=True, 
        label=" ",
        caption= " ",
        environment = "table*",
        )

    #erase last cline
    last_cline_starts = latex_code.find("\\cline", -100,-1)
    last_cline_ends = latex_code.find("\\bottomrule")
    latex_code = latex_code[:last_cline_starts] + latex_code[last_cline_ends:]

    latex_code = add_cline("NDCG", latex_code, df)
    latex_code = add_cline("FSat.*our", latex_code, df)
    latex_code = add_cline("IBO.*our", latex_code, df)
    latex_code = add_cline("VoCD", latex_code, df)
        
    latex_code = latex_code.replace("\t","\\t")

    now = datetime.now()
    time = str(now.strftime("%Y-%m-%d_%H%M%S"))

    with open(f'base/out_{time}.tex', 'w') as f:
        print(latex_code, file=f)
```

```python

for i, data in enumerate(["Amazon_Luxury_Beauty","lastfm",
        "ml-1m",
        "book-crossing",
        "Amazon_Industrial_and_Scientific",
        "Amazon_Digital_Music"
        ]
        ):
    plt.figure(figsize=(15,12))

    forcorr = temp.reindex(measure_order, level=1)

    forcorr = forcorr.loc[data].dropna().T

    forcorr.columns = forcorr.columns.str.strip("@10")
    forcorr.columns = forcorr.columns\
                        .str.replace("_pred","-pred", regex=False)\
                        .str.replace("_true","-true", regex=False)\
                        .str.replace("_our","$ _{our}$")\
                        .str.replace("_ori","$ _{ori}$")

    forcorr.loc[:,forcorr.columns.str.contains("(Gini)|(VoCD)|(IWO)|(IAA)|(MME)|(II\-F)|(AI\-F)",regex=True)] = forcorr.loc[:,forcorr.columns.str.contains("(Gini)|(VoCD)|(IWO)|(OD)|(IAA)|(MME)|(II\-F)|(AI\-F)",regex=True)].apply(lambda x: -x)
    mask = np.triu(np.ones_like(forcorr.corr("kendall").round(2), dtype=np.bool))
    sns.heatmap(forcorr.corr("kendall").round(2), mask=mask, annot=True, square=True, vmin=-1, vmax=1, cmap="coolwarm_r", annot_kws={"size": 10})
    if data == "Amazon_Luxury_Beauty":
        data = "amazon-lb"
    elif data == "Amazon_Industrial_and_Scientific":
        data = "amazon-is"
    elif data == "book-crossing":
        data = "book-x"
    elif data == "Amazon_Digital_Music":
        data = "amazon-dm"
    plt.title(f"Correlation of measures at $k=10$ ({data.capitalize()})", size=15)
    plt.savefig(f'base/triangular_corr_plot_{data}.png')
    plt.tight_layout()
    plt.show()
    
```
